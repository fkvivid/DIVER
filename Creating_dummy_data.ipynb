[Problem 1] Creation of random numbers

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]] 
np.random.seed(0)
x, y = np.random.multivariate_normal(mean, cov, 500).T
a = x,y
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

[Problem 2] Visualization by scatter plot

```python
plt.scatter(x,y)
```

[Problem 3] Visualization by histogram

```python
plt.hist(x)
plt.xlim()
```

[Problem 4] Addition of data

```python
mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]] 
np.random.seed(0)
x1, y1 = np.random.multivariate_normal(mean, cov, 500).T
b = x1,y1
plt.scatter(x,y)
plt.scatter(x1,y1)
```

[Problem 5] Data combination

```python
c = np.concatenate((a, b), axis=1)
#c = np.vstack((a,b))
print(c.shape)
```

[Problem 6] Labeling

```python
d = np.ones((1,1000))
print(d.shape)
e = np.vstack((c,d))
print(e.shape)
```
